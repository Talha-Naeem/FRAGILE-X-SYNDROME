**Fragile X syndrome (FXS) is a genetic disease that occurs due to the mutation of  CGG  TriNucleotide Repeats. FXS consider as the most common inherited cause of  intellectual   disability   and  the  second  most  prevalent  cause  after Down   Syndrome.  TriNucleotide  Repeats   is   a    type   of   mutation  in   which   repeats  of   three   Nucleotides    increase  the  number  of  copies  before  they   reach  the  threshold   over  which  they  become  unstable.It is important to understand the mutated FMR1 gene which is the cause of FXS.**

**A strategy is made with the aid of Bioinformatics expertise in which we developed a pipeline by using various python libraries and predictive algorithms. The pipeline named as TNR Analyzer is able to find the positions of trinucleotide repeats in the mutated FXS sequences.**

**CNNPromotor_e is used to predict promoters and FGENES-M predicts the exons, proteins, and also find the GC content of mutated FXS sequence. We have combined the outputs of these predictive algorithms with our pipeline.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("sequence1.fasta.csv")
```

```python
data1 = data1.rename({'>NM_001146070.2 Homo sapiens tudor domain containing 3 (TDRD3)': 'seq'}, axis='columns')
```

```python
data1
```

```python
data1.insert(1, "Occur", data1['seq'].str.count('CGG'), True)
data1
```

```python
#With loop
data1['CGG'] = data1['seq'].str.count('CGG')
print('Occur\t','CGG')
for i in data1['CGG']:
    print(i,'\t',i*'CGG')
```

```python
cnt=data1['seq'].str.count('CGG')
a=[]
for i in cnt:
    a.append(i*'CGG')
data1['CGG']=a
```

```python
seq1 = pd.DataFrame(data1, columns=['seq', 'Occur','CGG'])
```

**This  output shows that we calculate the total CGG occurence in  sequence in number and alphabetical form**

```python
seq1
```

```python
Total = data1['Occur'].sum()
print('Total Occurence of CGG: ',Total)
```

```python
plt.bar(data1.index,data1['Occur'])
plt.xlabel(" Location of CGG ",color='blue')
plt.ylabel("CGG Occurence",color='red')
plt.title('Mutation Analysis of CGG')
plt.legend(["CGG Occur"])
plt.show()
```
